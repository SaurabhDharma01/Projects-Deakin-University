## Exploring NHANES Body Measurements Dataset

### Introduction

In this Jupyter Notebook, we will explore body measurement data from the National Health and Nutrition Examination Survey (NHANES) dataset. The dataset contains body measurements of adult males and females. We will download two excerpts from the dataset, specifically:

1. `nhanes_adult_male_bmx_2020.csv`
2. `nhanes_adult_female_bmx_2020.csv`

We will use the `numpy` library to read and process the data, creating two matrices named `male` and `female`. Each matrix consists of seven columns representing different body measurements:

1. Weight (kg)
2. Standing height (cm)
3. Upper arm length (cm)
4. Upper leg length (cm)
5. Arm circumference (cm)
6. Hip circumference (cm)
7. Waist circumference (cm)

Additionally, we will calculate and add an eighth column to both matrices, storing the body mass indices (BMI) of the participants.

*We have already made a copy of csv file where we edited first few line that described the data and stored for further analysis using python.*

Let's start by downloading and processing the data.

**Data Loading**

```python
import numpy as np
from io import StringIO

# Define data
male_path = 'nhanes_adult_male_bmx_2020 - Copy.csv'
female_path = 'nhanes_adult_female_bmx_2020 - Copy.csv'

# Read data into NumPy matrices
with open(male_path, 'r') as male_file:
    male_data = male_file.read()
    male = np.genfromtxt(StringIO(male_data), delimiter=',', skip_header=1)

with open(female_path, 'r') as female_file:
    female_data = female_file.read()
    female = np.genfromtxt(StringIO(female_data), delimiter=',', skip_header=1)

# Display the matrices
print("Male Matrix:")
print(male)

print("\nFemale Matrix:")
print(female)

# Display shape of data
print("\nShape of Male Matrix:")
print(male.shape)

print("\nShape of Female Matrix:")
print(female.shape)
```

Both matrices have 7 columns and male data has 4080 entries and female has 4220 entries.

**Assignment of names**

Let us assingn names to the columns.

```python
column_names = [
    "Weight (kg)",
    "Standing Height (cm)",
    "Upper Arm Length (cm)",
    "Upper Leg Length (cm)",
    "Arm Circumference (cm)",
    "Hip Circumference (cm)",
    "Waist Circumference (cm)"
]
```

```python
# Print column names along with corresponding data for male
for i, name in enumerate(column_names):
    print(f"{name}: {male[:, i]}")  # Male data matrix
```

```python
# Print column names along with corresponding data for females
for i, name in enumerate(column_names):
    print(f"{name}: {female[:, i]}")  # Female data matrix
```

**BMI Calculation**

We will now create another column to store calculated BMI.

```python
# Calculate BMI for males and add a new column
male_bmi = male[:, 0] / ((male[:, 1] / 100) ** 2)
male = np.column_stack((male, male_bmi))

# Calculate BMI for females and add a new column
female_bmi = female[:, 0] / ((female[:, 1] / 100) ** 2)
female = np.column_stack((female, female_bmi))

# Print the updated matrices with BMI column
print("Updated Male Matrix with BMI:")
print(male)

print("\nUpdated Female Matrix with BMI:")
print(female)
```

**Visualisation of data**
- Histograms

Let us visualise data using a histograms to check the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'male_with_bmi' and 'female_with_bmi' are your matrices with BMI included

# Set the number of bins for the histograms
num_bins = 20

# Create subplots
plt.figure(figsize=(10, 8))

# Plot histogram for male BMIs (top subfigure)
plt.subplot(2, 1, 1)
plt.hist(male[:, -1], bins=num_bins, color='blue', edgecolor='black')
plt.title('Histogram of Male BMIs')
plt.xlabel('BMI')
plt.ylabel('Frequency')

# Plot histogram for female BMIs (bottom subfigure)
plt.subplot(2, 1, 2)
plt.hist(female[:, -1], bins=num_bins, color='pink', edgecolor='black')
plt.title('Histogram of Female BMIs')
plt.xlabel('BMI')
plt.ylabel('Frequency')

# Set identical x-axis limits for both subplots
plt.xlim(left=min(np.min(male[:, -1]), np.min(female[:, -1])),
         right=max(np.max(male[:, -1]), np.max(female[:, -1])))

# Adjust layout for better visualization
plt.tight_layout()

# Show the plot
plt.show()
```

- Box plot

Let us also take a look at the outliers and central tendencies through a box plot.

```python
import matplotlib.pyplot as plt

# Combine male and female BMIs into a list
bmi_data = [male[:, -1], female[:, -1]]

# Create a box-and-whisker plot with color
plt.figure(figsize=(8, 6))

# Set patch_artist=True for color
box = plt.boxplot(bmi_data, labels=['Male', 'Female'], patch_artist=True)

# Add color to the boxes
colors = ['blue', 'pink']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Set labels and title
plt.title('Box-and-Whisker Plot of Male and Female BMIs')
plt.xlabel('Gender')
plt.ylabel('BMI')

# Show the plot
plt.show()
```

**Computing the basic numerical aggregates of the male and female BMIs**

```python
import numpy as np
from scipy.stats import skew, kurtosis

# Male BMI stats
mean_male = np.mean(male[:, -1])
median_male = np.median(male[:, -1])
std_male = np.std(male[:, -1])
IQR_male = np.percentile(male[:, -1], 75) - np.percentile(male[:, -1], 25)
skewness_male = skew(male[:, -1])
kurtosis_male = kurtosis(male[:, -1])

# Female BMI stats
mean_female = np.mean(female[:, -1])
median_female = np.median(female[:, -1])
std_female = np.std(female[:, -1])
IQR_female = np.percentile(female[:, -1], 75) - np.percentile(female[:, -1], 25)
skewness_female = skew(female[:, -1])
kurtosis_female = kurtosis(female[:, -1])

# Report the results
print("Male BMI Stats:")
print(f"Mean: {mean_male:.2f}")
print(f"Median: {median_male:.2f}")
print(f"Standard Deviation: {std_male:.2f}")
print(f"IQR: {IQR_male:.2f}")
print(f"Skewness: {skewness_male:.2f}")
print(f"Kurtosis: {kurtosis_male:.2f}\n")

print("Female BMI Stats:")
print(f"Mean: {mean_female:.2f}")
print(f"Median: {median_female:.2f}")
print(f"Standard Deviation: {std_female:.2f}")
print(f"IQR: {IQR_female:.2f}")
print(f"Skewness: {skewness_female:.2f}")
print(f"Kurtosis: {kurtosis_female:.2f}")
```

### Analysing above findings

**Male BMI Distribution:**

- Central Tendency: The mean BMI for males is 29.14, and the median is 28.26, with the mean slightly higher than the median, suggesting a right-skewed distribution.

- Dispersion: The standard deviation is 6.31, and the interquartile range (IQR) is 7.73. Both values indicate moderate dispersion in male BMI, capturing the spread of data.

- Shape: Skewness is 0.97, indicating a rightward (positive) skew, and kurtosis is 1.76, suggesting heavier tails and a more peaked distribution than normal.

**Female BMI Distribution:**

- Central Tendency: The mean BMI for females is 30.10, and the median is 28.88, with a right-skewed distribution similar to males.

- Dispersion: The standard deviation is 7.76, and the IQR is 10.01, indicating higher dispersion and a broader spread of BMI values compared to males.

- Shape: Skewness is 0.92, showing a rightward (positive) skew similar to males. Kurtosis is 1.10, indicating a distribution somewhat less peaked than the male distribution.

**Comparison:**

- Both distributions are right-skewed, indicating a prevalence of lower BMI values with a tail towards higher values.

- The female distribution has a slightly higher mean, higher standard deviation, and a broader IQR, indicating greater variability in female BMI values compared to males.

- Positive skewness in both distributions implies a longer right tail, suggesting the presence of individuals with higher BMI values.

**Summary:**

While both distributions share right-skewed characteristics, the female BMI distribution stands out with a slightly higher mean, greater dispersion, and a broader spread of values compared to the male distribution.

### Analysing relationship between features in male data

- Scatter plot

We create a scatter plot for "Height", "Weight", "Waist Circumference", "Hip Circumference" & "BMI" to analyse

```python
# Select the relevant columns
selected_columns = [1, 0, -2, -3, -1]  # Indices for height, weight, waist, hip, BMI
male_selected = male[:, selected_columns]

# Get the names of the selected columns
column_names = ["Height", "Weight", "Waist Circumference", "Hip Circumference", "BMI"]

# Create a scatterplot matrix
num_columns = len(selected_columns)

fig, axes = plt.subplots(nrows=num_columns, ncols=num_columns, figsize=(15, 15))

for i in range(num_columns):
    for j in range(num_columns):
        if i == j:
            axes[i, j].hist(male_selected[:, i], bins=20, color='skyblue', edgecolor='black')
            axes[i, j].set_title(column_names[i])
        else:
            scatter = axes[i, j].scatter(
                male_selected[:, j],
                male_selected[:, i],
                c=male_selected[:, -1],  # Use BMI values for color gradient
                cmap='viridis',  # Choose a colormap
                alpha=0.7,
            )
            axes[i, j].set_xlabel(column_names[j])
            axes[i, j].set_ylabel(column_names[i])
            plt.colorbar(scatter, ax=axes[i, j])  # Add colorbar for better interpretation

plt.suptitle('Scatterplot Matrix of Male Heights, Weights, Waist Circumferences, Hip Circumferences, and BMIs', y=1)
plt.tight_layout()
plt.show()
```

**Computing Pearsons and Spearmans correlation**

- corrheatmap

We will now create a corheatmap to visualise relationship between all the features.

```python
import seaborn as sns
from scipy.stats import pearsonr, spearmanr

# Compute Pearson's correlation coefficients
pearson_corr = np.corrcoef(male_selected, rowvar=False)

# Compute Spearman's correlation coefficients
spearman_corr = np.zeros_like(pearson_corr)
for i in range(male_selected.shape[1]):
    for j in range(male_selected.shape[1]):
        spearman_corr[i, j], _ = spearmanr(male_selected[:, i], male_selected[:, j])

# Visualize correlation matrices with heatmaps
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Pearson's correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt='.2f', xticklabels=column_names, yticklabels=column_names, ax=axes[0])
axes[0].set_title("Pearson's Correlation Coefficients")

# Spearman's correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt='.2f', xticklabels=column_names, yticklabels=column_names, ax=axes[1])
axes[1].set_title("Spearman's Correlation Coefficients")

plt.suptitle('Correlation Heatmaps for Male Heights, Weights, Waist Circumferences, Hip Circumferences, and BMIs', y=1.02)
plt.tight_layout()
plt.show()
```
